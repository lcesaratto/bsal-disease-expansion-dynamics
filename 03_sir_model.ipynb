```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Exponential decay (example from Scipy)

```python
def exponential_decay(t, y):
    # this is dy/dt
    return -0.5 * y

sol = solve_ivp(fun=exponential_decay, t_span=[0, 10], y0=[2, 4, 8])

plt.figure()
sns.lineplot(x=sol.t, y=sol.y[0])
sns.lineplot(x=sol.t, y=sol.y[1])
sns.lineplot(x=sol.t, y=sol.y[2])

plt.plot()
```

### SIR model

```python
def sir_model(t, y, vu, gamma, mu, beta):
    """
    vu: birth rate per individual
    gamma: recovery rate per individual (includes natural deaths in R)
    mu: general natural death rate per individual
    beta: infection transmission rate
    """
    S, I, R = y
    # print(S, I, R)
    N = S + R + I
    dS_dt = vu * N - beta * S * I - mu * S
    dI_dt = beta * S * I - gamma * I - mu * I
    dR_dt = gamma * I - mu * R
    return [dS_dt, dI_dt, dR_dt]

sol = solve_ivp(fun=sir_model, t_span=[0, 15], t_eval=np.linspace(0, 15, 1000), y0=[45, 5, 0], args=(0, 0.3, 0.05, 0.12))

fig, ax = plt.subplots(1,2, figsize=(10,3))
sns.lineplot(x=sol.t[:100], y=sol.y[0][:100], label="S", ax=ax[0])
sns.lineplot(x=sol.t[:100], y=sol.y[1][:100], label="I", ax=ax[0])
sns.lineplot(x=sol.t[:100], y=sol.y[2][:100], label="R", ax=ax[0])
sns.lineplot(x=sol.t, y=sol.y[0], label="S", ax=ax[1])
sns.lineplot(x=sol.t, y=sol.y[1], label="I", ax=ax[1])
sns.lineplot(x=sol.t, y=sol.y[2], label="R", ax=ax[1])
plt.plot()
```
